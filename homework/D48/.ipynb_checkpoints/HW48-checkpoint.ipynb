```python
import numpy as np 
import pandas as pd
from sklearn.metrics import accuracy_score,roc_auc_score,roc_curve,auc,confusion_matrix
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt



train_data = pd.read_csv('train.csv',header = None)
train_labels = pd.read_csv('trainLabels.csv',header = None)
test_data =  pd.read_csv('test.csv',header = None)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data, train_labels, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
predicted= lr.predict(X_test)
print('LogisticRegression: ',accuracy_score(y_test, predicted))
print('LogisticRegression AUC: ',roc_auc_score(y_test, predicted))
fpr,tpr,threshold = roc_curve(y_test.values, predicted) ###計算真正率和假正率
roc_auc = auc(fpr,tpr) ###计算auc的值
print('LogisticRegression AUC: ',roc_auc)

```

```python
# Compute ROC curve and ROC area for each class
fpr,tpr,threshold = roc_curve(y_test.values, predicted) ###計算真正率和假正率
roc_auc = auc(fpr,tpr) ###计算auc的值

plt.figure()
lw = 2
plt.figure(figsize=(10,10))
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.3f)' % roc_auc) ###假正率為橫坐標，真正率為縱坐標做曲線
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
def modelfit(alg, dtrain, labrain, performCV=True, printFeatureImportance=True, cv_folds=5):
    #Fit the algorithm on the data
    alg.fit(dtrain, labrain)
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain)
    dtrain_predprob = alg.predict_proba(dtrain)[:,1]
    
    #Perform cross-validation:
    if performCV:
        cv_score = cross_val_score(alg,dtrain, labrain, cv=cv_folds, scoring='roc_auc')
    
    #Print model report:
    print("\nModel Report")
    print( "Accuracy : %.4g" % accuracy_score(labrain.values, dtrain_predictions))
    print( "AUC Score (Train): %f" % roc_auc_score(labrain, dtrain_predprob))
    
    if performCV:
        print("CV Score : Mean - %.7g | Std - %.7g | Min - %.7g | Max - %.7g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
        
    #Print Feature Importance:
    if printFeatureImportance:
        feat_imp = pd.Series(alg.feature_importances_, dtrain.columns).sort_values(ascending=False)
        feat_imp.plot(kind='bar', title='Feature Importances')
        plt.ylabel('Feature Importance Score')
    
    fdf = pd.DataFrame({'feature':dtrain.columns.tolist(), 'feature_importance':alg.feature_importances_.tolist()})
    fdf = fdf[fdf['feature_importance'] > 0.01]
    lessfdf = pd.DataFrame()

    for n in fdf['feature'].values:
        lessfdf[n] = dtrain[n]
    
    return (alg,lessfdf, fdf)
    
    

```

```python
modell = GradientBoostingClassifier()
lessfobj = modelfit(modell, X_train, y_train)
# xx = modelfit(modell, lessfobj[1], y_train)
```
