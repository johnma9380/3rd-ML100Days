```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
from scipy.stats import mode

mode_get = mode(app_train[~app_train['HOUSETYPE_MODE'].isnull()]['HOUSETYPE_MODE'])
print(mode_get)
print(mode_get[0][0])   # 將值拿出來

app_train.loc[app_train['HOUSETYPE_MODE'].isnull(), 'HOUSETYPE_MODE'] = mode_get[0][0]

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
app_train['YEARS_BIRTH'] = app_train['DAYS_BIRTH']/365
```

```python
app_train['HOUSETYPE_MODE'].unique()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
import seaborn as sns # 另一個繪圖-樣式套件

unique_house_type = app_train['HOUSETYPE_MODE'].unique()

nrows = len(unique_house_type)
ncols = nrows // 2

plt.clf() 
plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    sns.distplot(app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i], 'YEARS_BIRTH'], 
                 label = str(unique_house_type[i]), hist = True)
plt.show()    
```
