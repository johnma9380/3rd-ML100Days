```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
# train_Y = df_train['SalePrice']

df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
import seaborn as sns
import matplotlib.pyplot as plt

cdf = df.sort_values(by='1stFlrSF',ascending=True).reset_index(drop=True)
Q1 = cdf['1stFlrSF'].quantile(0.25)
Q3 = cdf['1stFlrSF'].quantile(0.75)
IQR = Q3 - Q1
df['1stFlrSF'] = df['1stFlrSF'].clip( (Q1 - 1.5 * IQR), (Q3 + 1.5 * IQR))
sns.regplot(x = df['1stFlrSF'], y=train_Y)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()

## 效果沒有截尾好
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
import matplotlib.pyplot as plt

df['1stFlrSF'].value_counts()
plt.hist(df['1stFlrSF'])
plt.show()

plt.boxplot(df['1stFlrSF'])
plt.show()


cdf = df.sort_values(by='1stFlrSF',ascending=True).reset_index(drop=True)

plt.plot(np.log(cdf['1stFlrSF']), np.log(cdf.index))
plt.xlabel('Value')
plt.ylabel('ECDF')
# plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
# plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

Q1 = cdf['1stFlrSF'].quantile(0.25)
Q3 = cdf['1stFlrSF'].quantile(0.75)
IQR = Q3 - Q1
keep_indexs = ((df['1stFlrSF'] > (Q1 - 3 * IQR)) & (df['1stFlrSF'] < (Q3 + 3 * IQR)))

df2 = df[keep_indexs]
train_Y2 = train_Y[keep_indexs]
sns.regplot(x = df2['1stFlrSF'], y=train_Y2)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df2)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y2, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np


keep_indexs = np.abs(stats.zscore(df['1stFlrSF'])) <4

df2 = df[keep_indexs]
train_Y2 = train_Y[keep_indexs]
sns.regplot(x = df2['1stFlrSF'], y=train_Y2)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df2)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y2, cv=5).mean()
```
